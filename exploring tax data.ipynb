```python
from docx import Document
```

```python
def read_table_from_docx(docx_path):
    doc = Document(docx_path)
    tables = doc.tables

    data = {'Column1': [], 'Column2': [], 'Column3': []}  # Adjust column names accordingly

    for table in tables:
        for row in table.rows:
            columns = row.cells
            data['Column1'].append(columns[0].text)
            data['Column2'].append(columns[1].text)
            data['Column3'].append(columns[2].text)

    df = pd.DataFrame(data)
    return df
```

```python
irs_table = r'data/irs_data/2020 irs table.docx'
df = read_table_from_docx(irs_table)
```
