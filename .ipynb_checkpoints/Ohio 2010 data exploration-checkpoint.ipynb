```python
#get list of all candidates
def get_candidates(data,delimiter='-'):
    """this function will get candidate names when the format is 'type of race - district - candidate name'"""
    cand_split = [cand.split(delimiter) for cand in data]
    cand_names = [cand[-1] for cand in cand_split]
    cand_names = [cand.strip() for cand in cand_names]
    return cand_names
```

```python
#get list of all districts
def get_districts(data,delimiter='-'):
    """this function will get district names when the format is 'type of race - district - candidate name'"""
    data_split = [dist.split(delimiter) for dist in data]
    district_names = [dist[-2] for dist in data_split]
    district_names = [dist.strip() for dist in district_names]
    return district_names
```

```python
def remove_middle_names(name):
    # Split the name by spaces
    parts = name.split(' ')
    
    # If the item has more than one space, keep only the first two parts
    if len(parts) > 2:
        return ' '.join(parts[:2])
    else:
        return name
```

```python
#remove middele names from this df
house_cand_2010_simple['CANDIDATE NAME (Last, First)'] = house_cand_2010_simple['CANDIDATE NAME (Last, First)'].apply(lambda x : remove_middle_names(x))
```
